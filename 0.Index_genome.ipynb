# Global variables

```bash
RefDir=/Ref
RefFasta=${RefDir}/Ppatens_870_V6.fa
RefGff3=${RefDir}/Ppatens_870_v6.1.gene.gff3
Species=PP
Retrieve_seq=/path_to_annovar/annovar/retrieve_seq_from_fasta.pl

AnnoVarDB=${RefDir}/Anovar_db
mkdir -p $AnnoVarDB
```

# Index reference genome

```bash
# Build reference indexes
samtools faidx $RefFasta
picard CreateSequenceDictionary -R $RefFasta
bwa index $RefFasta

# Generate refGene format for ANNOVAR
gff3ToGenePred $RefGff3 $AnnoVarDB/${Species}_refGene.txt

# Retrieve mRNA sequences
$Retrieve_seq \
    --format refGene \
    --seqfile $RefFasta \
    $AnnoVarDB/${Species}_refGene.txt \
    --out $AnnoVarDB/${Species}_refGeneMrna.fa
```
